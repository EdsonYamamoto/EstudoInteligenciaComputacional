### Exercício 3: Utilize o método de busca de parâmetros e mostre a curva de resultados para redes neurais baseado nos resultados obtidos no exercício anterior.

```python
import numpy as np
```

```python

# importação para retirada dos avisos
import warnings
warnings.filterwarnings("ignore")

# importação da biblioteca do matplot lib
import matplotlib.pyplot as plt

# importação da biblioteca para pegar a curva de aprendizado
from sklearn.model_selection import learning_curve
```

```python
#Criação da função que carrega os dados obtidos 
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 10)):
    #definição dos parametros do desenho
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Quantidade de examplos para treino")
    plt.ylabel("Score")
    
    #definição dos parametro para curva de aprendizado
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    #extração de atributos para a média de treino
    train_scores_mean = np.mean(train_scores, axis=1)
    #extração de atributos para o desvio de treino
    train_scores_std = np.std(train_scores, axis=1)
    #extração de atributos para a média de teste
    test_scores_mean = np.mean(test_scores, axis=1)
    #extração de atributos para o desvio de teste
    test_scores_std = np.std(test_scores, axis=1)

    #plotagem do grafico

    plt.grid()
    #plotagem dos dados no grafico
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")

    #adição dos atributos e do tipo de grafico
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    #adição dos atributos e do tipo de grafico
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
#plotagem do hold out
plot_learning_curve(holdOut, 'Learning Curve', X, y, cv=5, n_jobs=4)
```
